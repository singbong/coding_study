```python
import pandas as pd
table= pd.crosstab(index= death, colnames=["질병"], columns="도수")
#table['relative ratio']= np.round(table.도수/sum(table.도수),2)
table.index = ['감염','각종암','순환기','호흡기','소화기','사고사','비뇨기','정신병','노환','신경계']
print(table)
```

## 막대 그래프

```python
import matplotlib.pyplot as plt
table.plot(kind='bar', color='blue',legend=False)
plt.xlabel('사망원인')
plt.ylabel('빈도수')
plt.title("사망원인에 따른 막대 그래프")
plt.xticks(rotation=0)
plt.show

plt.savefig('./plot')
```

### 원형 그래프

```python
plt.figure()
index = ['감염','각종암','순환기','호흡기','소화기','사고사','비뇨기','정신병','노환','신경계']
plt.pie(table['도수'], labels= index)
plt.rc('font', size=9)
plt.title("사망원인에 대한 원형 그래프")
plt.show
```
